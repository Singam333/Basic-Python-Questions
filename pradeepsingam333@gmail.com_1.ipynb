```python
import operator
def CountFrequency(my_list): 
  
    # Creating an empty dictionary  
    freq = {} 
    for item in my_list: 
        if (item in freq): 
            freq[item] += 1
        else: 
            freq[item] = 1
    freq = sorted(freq.items(), key=operator.itemgetter(1))
    freq=dict(freq)
  
    for key, value in freq.items(): 
        print ("% d : % d"%(key, value))
```

```python
Red= [(1,1),(2,1),(4,2),(2,4), (-1,4)]
Blue= [(-2,-1),(-1,-2),(-3,-2),(-3,-1),(1,-3)]
Lines=["1x+1y+0","1x-1y+0","1x+0y-3","0x+1y-0.5"]
def split(word): 
    return [char for char in word]  

for i in Lines:
    l=[]
    l1=[]
    m=[]
    m1=[]
    s1=split(i)
    x=int(s1[0])
    if(s1[2]=='+'):
        y=int(s1[3])
    else:
        y=-1*(int(s1[3]))
    if(s1[5]=='+'):
        intercept=int(s1[6])
    else:
        intercept=-1*(int(s1[6]))
    for j in(Red):
        l.append(x*j[0]+y*j[1]+intercept)
    for k in(Blue):
        l1.append(x*k[0]+y*k[1]+intercept)
    for a in l:
        if(a>0):
            m.append(True)
        else:
            m.append(False)
    for a in l1:
        if(a<0):
            m1.append(True)
        else:
            m1.append(False)
    n=all(m)
    n1=all(m1)
    if(n==True and n1==True):
        print("yes")

    else:
        print("no")
```

```python
def fun(x, a, b):
    if a == -1:
        v = float(x[b])/(b+1)
        for i in range(a+1,b+1):
            x[i] = v
    elif b == -1:
        v = float(x[a])/(len(x)-a)
        for i in range(a, len(x)):
            x[i] = v
    else:
        v = (float(x[a])+float(x[b]))/(b-a+1)
        for i in range(a,b+1):
            x[i] = v
    return x
def replace(text):
    # Create array from the string
    x = text.replace(" ","").split(",")
    # Get all the pairs of indices having number
    y = [i for i, v in enumerate(x) if v != '_']
    # Starting with _ ?
    if y[0] != 0:
        y = [-1] + y
    # Ending with _ ?
    if y[-1] != len(x)-1:
        y = y + [-1]    
    # run over all the pairs
    for (a, b) in zip(y[:-1], y[1:]): 
        fun(x,a,b)          
    return x

tests = [
    "_,_,_,24",
    "40,_,_,_,60",
    "80,_,_,_,_",
     "_,_,30,_,_,_,50,_,_"]

for i in tests:
    print (replace(i))
```
